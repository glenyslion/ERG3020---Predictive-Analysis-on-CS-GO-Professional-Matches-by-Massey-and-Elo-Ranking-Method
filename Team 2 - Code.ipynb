# Massey Method

```python
def MasseyRating(df):
  list_team_1 = list(df.team_1.unique())
  list_team_2 = list(df.team_2.unique())
  unique_team = list_team_1 + list(set(list_team_2) - set(list_team_1))
  B_matrix = np.zeros((df.shape[0], len(unique_team)))

  for i in range(df.shape[0]):
    if(df.iloc[i,5] == 1): # df.iloc[i,5] -> check the match winner of each row
      pos_in_list_unique_team_winner = unique_team.index(df.iloc[i,0])
      pos_in_list_unique_team_loser = unique_team.index(df.iloc[i,1])
      B_matrix[i, pos_in_list_unique_team_winner] = 1
      B_matrix[i, pos_in_list_unique_team_loser] = -1
    else:
      pos_in_list_unique_team_winner = unique_team.index(df.iloc[i,1])
      pos_in_list_unique_team_loser = unique_team.index(df.iloc[i,0])
      B_matrix[i, pos_in_list_unique_team_winner] = 1
      B_matrix[i, pos_in_list_unique_team_loser] = -1

  # In here, the B matrix is not linearly independent. To make it linearly dependent and can be calculated using least square, we add the last row of the B matrix to a row of all ones
  # and the corresponding entry in the score difference vector to zero.
  last_row = np.ones((1, len(unique_team)))
  B_matrix = np.concatenate((B_matrix, last_row), axis=0)
  # print(B_matrix)

  # v matrix
  v = np.array(df['score_diff'])
  v = np.concatenate([v, np.zeros(1)])
  # print(v)

  # calculate the r = (B^TB)^-1 B^tv
  r = np.matmul(np.linalg.pinv(B_matrix),v)
  # print(r)

  # the r value here is in order with the unique team
  massey_rating = pd.DataFrame(list(zip(unique_team, r)), columns = ['Team', 'Massey Rating'])
  massey_rating = massey_rating.sort_values(by = ['Massey Rating'], ascending = False)
  return massey_rating
```

```python
def accuracy(dict_df):
  # Test: For team without history, it wont be counted as prediction (skipped). 
  # Note: Tested data will get augmented into trained model.

  # Initialize variables for accuracy calculation
  correct_predictions = 0
  total_predictions = 0

  for index, row in test_df.iterrows():
    
    team = row['team_1']
    team_score = row['result_1']
    opponent = row['team_2']
    opponent_score = row['result_2']

    # check who won
    if team_score > opponent_score:
        winner = team
        loser = opponent
    else:
        winner = opponent
        loser = team

    
    # Check if both teams have Elo ratings
    if (team in dict_df) and (opponent in dict_df): # Available

          # Get current ratings based on Elo ratings
          team_rating = dict_df[team]
          # print(team_rating)
          opponent_rating = dict_df[opponent]
          # print(opponent_rating)
          
          # Predict the winner based on Elo ratings
          predicted_winner = team if team_rating > opponent_rating else opponent

          # Check if the prediction matches the actual result
          if (predicted_winner == winner):
              correct_predictions += 1
          total_predictions += 1

          

  # Calculate accuracy
  accuracy = correct_predictions / total_predictions
  # print(total_predictions)

  # Print the accuracy result
  print(f"Accuracy: {accuracy * 100}%")
```

# Elo Model Training

```python
# function to update elo ratings based on result
def update_ratings(winner_rating, loser_rating, k):
    
    # get expected scores for each team
    E_winner = 1/(1+(10 ** (-(winner_rating-loser_rating)/400)))
    E_loser = 1/(1+(10 ** (-(loser_rating-winner_rating)/400)))
    
    # Initialize actual scores
    S_winner = 1
    S_loser = 0
    
    # update elo ratings
    new_winner_rating = winner_rating + (k * (S_winner - E_winner))
    new_loser_rating = loser_rating + (k * (S_loser - E_loser))
    
    return [new_winner_rating, new_loser_rating]
```

```python
def check_accuracy(k):
    correct_predictions = 0
    total_predictions = 0

    for index, row in test_df.iterrows():
        
        team = row['team_1']
        team_score = row['result_1']
        opponent = row['team_2']
        opponent_score = row['result_2']

        # check who won
        if team_score > opponent_score:
            winner = team
            loser = opponent
        else:
            winner = opponent
            loser = team

        # Check if both teams have Elo ratings
        if (team in ratings_dict) and (opponent in ratings_dict): # Available

              # Get current ratings based on Elo ratings
              team_rating = ratings_dict[team]
              opponent_rating = ratings_dict[opponent]
              
              # Predict the winner based on Elo ratings
              predicted_winner = team if team_rating > opponent_rating else opponent

              # Check if the prediction matches the actual result
              if (predicted_winner == winner):
                  correct_predictions += 1
              total_predictions += 1

              # Update the ratings for the next test(s)
              ratings_dict[winner] = update_ratings(ratings_dict[winner], ratings_dict[loser], k)[0]
              ratings_dict[loser] = update_ratings(ratings_dict[winner], ratings_dict[loser], k)[1]

        else:   # New team
              if (team not in ratings_dict):
                  ratings_dict[team] = 1500
              if (opponent not in ratings_dict):
                  ratings_dict[opponent] = 1500
              
              # Update ratings
              ratings_dict[winner] = update_ratings(ratings_dict[winner], ratings_dict[loser], k)[0]
              ratings_dict[loser] = update_ratings(ratings_dict[winner], ratings_dict[loser], k)[1]
              
    return correct_predictions, total_predictions
```

```python
# x, y initialization for plotting k-factor accuracy
x = []
y = []

for k in range(0, 101, 1):
    x.append(k)

    # Initialize Elo ratings for all teams in trained set
    team_names = set(train_df['team_1']).union(set(train_df['team_2']))
    ratings_dict = {team: 1500 for team in team_names}

    # TRAIN DATA
    for index, row in train_df.iterrows():
        
        team = row['team_1']
        team_score = row['result_1']
        opponent = row['team_2']
        opponent_score = row['result_2']
        
        # check who won
        if team_score > opponent_score:
            winner = team
            loser = opponent
        else:
            winner = opponent
            loser = team
            
        # get current ratings
        winner_rating = ratings_dict[winner]
        loser_rating = ratings_dict[loser]
        
        # update the ratings
        ratings_dict[winner] = update_ratings(winner_rating, loser_rating, k)[0]
        ratings_dict[loser] = update_ratings(winner_rating, loser_rating, k)[1]

    # ACCURACY TEST
    correct_predictions, total_predictions = check_accuracy(k)
    accuracy = correct_predictions / total_predictions
    y.append(accuracy)

    # Print the accuracy result
    # print(f"K: {k}, Accuracy: {accuracy * 100}%")
```

```python
# Make a graph for finding the best k-factor (between 0 and 100)
plt.plot(x, y)
plt.xlabel('K-factor')
plt.ylabel('Accuracy rate')
plt.title("Finding the best k-factor between 0 and 100")
plt.show()

print("The maximum accuracy rate is", max(y), "when", y.index(max(y)))
```
